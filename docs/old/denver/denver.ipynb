```python
import pandas as pd
import os
```

```python
ROOT_PATH = "../data/denver_costar/"
file_paths = []

for file in os.listdir(ROOT_PATH):
    file_paths.append((ROOT_PATH + file))

file_paths[:2]
```

```python
dfs = []

for file_path in file_paths:
    df = pd.read_excel(file_path)
    if "zipcode" in df.columns:
        df = df.drop(columns="zipcode")

    df["Zipcode"] = file_path.split("/")[-1][:-5]
    dfs.append(df)
```

```python
final_df = pd.concat(dfs)
```

```python
final_df.sample(5)
```

```python
final_df.to_csv('../output/Denver_data.csv')
```
